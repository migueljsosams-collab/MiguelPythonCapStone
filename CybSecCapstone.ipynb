```python
# ==========================================
# Cybersecurity Incident Analysis
# ==========================================

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("nvb_cybersecurity_incidents.csv")

# -------------------------------------------------
# Function (required programming concept)
# Counts occurrences for a given column
# -------------------------------------------------

def count_categories(column_name):
    return df[column_name].value_counts()

attack_counts = count_categories("attack_type")

print("\nAttack Types")
print(attack_counts)

# -------------------------------------------------
# Loop (another required concept)
# -------------------------------------------------

print("\nAverage Response Time by Severity")

severity_levels = ["Low", "Medium", "High", "Critical"]

for level in severity_levels:
    avg = df[df["severity"] == level]["response_time_hrs"].mean()
    print(level, ":", round(avg,2), "hours")

# ===================================================
# Visualization 1
# Attack Type Frequency
# ===================================================

plt.figure(figsize=(8,5))
attack_counts.plot(kind="bar")

plt.title("Number of Cybersecurity Incidents by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ===================================================
# Visualization 2
# Average Response Time by Severity
# ===================================================

response = df.groupby("severity")["response_time_hrs"].mean()

plt.figure(figsize=(6,5))
response.plot(kind="bar")

plt.title("Average Response Time by Severity")
plt.xlabel("Severity")
plt.ylabel("Average Response Time (Hours)")
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()

# ===================================================
# Additional Analysis
# ===================================================

print("\nAverage Systems Affected by Severity")
print(df.groupby("severity")["systems_affected"].mean())

print("\nPercentage of Incidents with Data Loss")

data_loss = (
    df["data_loss_occurred"].value_counts(normalize=True) * 100
)

print(data_loss)

# ===================================================
# Written Summary
# ===================================================

print("""
Summary

The dataset contains 350 cybersecurity incidents.

The first chart shows which attack types occur most frequently.
The second chart compares the average response time for each
severity level. In general, higher severity incidents tend to
require longer response times than lower severity incidents.

The analysis also provides information about the average number
of systems affected and the percentage of incidents that resulted
in data loss. These findings can help organizations understand
where cybersecurity resources should be focused and improve
incident response planning.
""")
```
